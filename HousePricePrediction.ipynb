```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_california_housing
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
house_price_ds = fetch_california_housing()
```

```python
print(house_price_ds)
```

```python
house_price_dataframe = pd.DataFrame(house_price_ds.data, columns= house_price_ds.feature_names)
house_price_dataframe.head()
```

```python
house_price_dataframe['price'] = house_price_ds.target
house_price_dataframe.head()
```

```python
house_price_dataframe.shape
```

```python
house_price_dataframe.isnull().sum()
```

```python
house_price_dataframe.describe()
```

```python
correlation = house_price_dataframe.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, cmap='Blues', square=True, annot=True, annot_kws={'size' : 8}, fmt='.1f')
```

```python
X = house_price_dataframe.drop(['price'], axis=1)
Y = house_price_dataframe['price']
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = XGBRegressor()
```

```python
model.fit(X_train,Y_train)
```

```python
training_data_prediction = model.predict(X_train)
print(training_data_prediction)
```

```python
score_1 = metrics.r2_score(Y_train, training_data_prediction)
score_2 = metrics.mean_absolute_error(Y_train, training_data_prediction)
print("Training Data R Squared", score_1)
print("Training data Mean Absolute error", score_2)
```

```python
test_data_prediction = model.predict(X_test)
score_1 = metrics.r2_score(Y_test, test_data_prediction)
score_2 = metrics.mean_absolute_error(Y_test, test_data_prediction)
print("Testing Data R Squared", score_1)
print("Testing data Mean Absolute error", score_2)
```

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted label")
plt.title("Actual Prices vs Predicted Prices")
plt.show()
```
